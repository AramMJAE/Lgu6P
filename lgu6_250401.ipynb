#loc vs iloc
-코드비교

##loc

```python
#tips.loc[tips['day'] == 'Sun']
#tips.loc[행,열]
tips.loc[0:2, ['total_bill', 'tip', 'day']]
```

```python
#iloc
tips.iloc[0:2, [0, 1, 4]]
```

```python

#tips.loc[:,:]
tips.loc[tips['total_bill'] <= 11, :].reset_index(drop=True).head(1)
```

```python
tips.loc[tips['time'] == 'Dinner',:].reset_index(drop=True)
```

```python
tips.loc[(tips['time'] == 'Dinner') & (tips['total_bill'] <= 11), :]
#조건식 2개
#tips.loc[(조건식1) & (조건식2), :]
```

```python
result = tips.loc[tips['time'] == 'Dinner' , :]

result.loc[result['total_bill'] <= 11, :]
```
